```python
df['days'] = df['datetime'].dt.day # Seperate days from datetime and store it in a new column
```

```python
df['years'] = df['datetime'].dt.year # Seperate years from datetime and store it in a new column
```

```python
x = df.iloc[:,10:14].values # Take the values of date and time as input
```

```python
x # Display the input array
```

```python
y = df['load'] # Take the values of load as target or output 
```

```python
y # Display the output array
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=20) # Divide the dataset into training and
# testing, here 70% is training data and 30% is testing data.
```

```python
from sklearn.tree import DecisionTreeRegressor # Import the DecisionTree model Regressor
```

```python
reg = DecisionTreeRegressor(random_state=20) #Make the object of the regressor 
```

```python
reg.fit(x_train,y_train) # Fit the training and testing data in the model using fit method
```

```python
y_pred = reg.predict(x_test) # Predict the output using the predict method
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score # Import the regression score function
```

```python
r2_score(y_test,y_pred) # Calculate the accuracy of the model
```

```python
y_pred = reg.predict([[15,14,12,2020]]) # Predict for 14 dec 2020 when time is 15 minutes
```
